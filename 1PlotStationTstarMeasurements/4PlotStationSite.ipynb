```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker
import os
sitefile_P="./Input/SR_Pwave/"
sitefile_S="./Input/SR_Swave/"
station=["G01","G02","G03","G04","G05",'G06','G07',"G08","G09","G10","G12","G13","G14","G15","G16"]
fig,ax=plt.subplots(5,3,figsize=(15,20),sharex="col", sharey="row")
fontsize=18
for row in range(5):
    for col in range(3):
        i=row*3+col
        filenameP=sitefile_P+"SR.{}_.dat".format(station[i]);
        filenameS=sitefile_S+"SR.{}_.dat".format(station[i]);
        siteP=np.loadtxt(filenameP);siteS=np.loadtxt(filenameS)
        ax[row,col].plot(siteP[:,0],siteP[:,1],'-r',label="P-wave")
        ax[row,col].plot(siteS[:,0],siteS[:,1],'--k',label="S-wave")
        ax[row,col].annotate(station[i], xy=(0.05, 0.9), xycoords='axes fraction',fontsize=fontsize)
        ax[row,col].legend(loc=1,fontsize=fontsize)
        ax[row,col].set_xscale("log")
        if col==0:
            ax[row,col].set_ylabel("Amplitude")
        if row==4:
            ax[row,col].set_xlabel("Frequency (Hz)")
            ax[row,col].set_xticks([1,2,5,10,20])
            ax[row,col].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
plt.tight_layout(w_pad=0.2,h_pad=0.2)
plt.savefig("SiteResponse.jpg")
plt.show()
# plt.clf()
```
